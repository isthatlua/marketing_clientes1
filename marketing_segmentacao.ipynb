```python
import pandas as pd
df = pd.read_csv('planilhas dados/ifood_df.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head() 
df.info()
df.describe() 
```

```python
print(df['Z_CostContact'].value_counts())
print(df['Z_Revenue'].value_counts())                             
```

```python
df.groupby('Response')['Income'].mean()
```

```python
df = df.drop(columns=['Z_CostContact', 'Z_Revenue'])
print("Columns have been dropped! New info:")
df.info()
```

```python
df.groupby('Response')['MntTotal'].mean()
```

```python

plt.figure(figsize=(8, 6))
sns.barplot(x='Response', y='MntTotal', data=df)
plt.title('Average Total Spending by Campaign Response', fontsize=16)
plt.ylabel('Average Total Spent ($)', fontsize=12)
plt.xlabel('Campaign Response (0 = No, 1 = Yes)', fontsize=12)
plt.show()
```

```python
df['TotalKids'] = df['Kidhome'] + df['Teenhome']

df[['Kidhome', 'Teenhome', 'TotalKids']].head()
```

```python
# 1. MÃ©dia de filhos por resposta Ã  campanha
print("NÃºmero mÃ©dio de filhos por resposta:")
print(df.groupby('Response')['TotalKids'].mean())

print("\n" + "="*40 + "\n") # Isso Ã© sÃ³ pra separar os resultados e ficar bonito

# 2. Gasto mÃ©dio por nÃºmero de filhos
print("Gasto mÃ©dio TOTAL por nÃºmero de filhos:")
print(df.groupby('TotalKids')['MntTotal'].mean())
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='TotalKids', y='MntTotal', data=df)
plt.title('Gasto MÃ©dio por NÃºmero de Filhos', fontsize=16)
plt.xlabel('NÃºmero Total de Filhos em Casa', fontsize=12)
plt.ylabel('Gasto MÃ©dio Total ($)', fontsize=12)
plt.show()
```
